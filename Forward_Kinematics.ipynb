```python
import numpy as np
import sympy.geometry as sg
import matplotlib.pyplot as plt
import csv
```

```python
b = np.array([[15, 0], [-15, 0]])
l = np.array([40, 40])
m = np.array([50, 50])
x = []
y = []
Thete1 = []
Thete2 = []
x_all = []
y_all = []
Thete1_all = []
Thete2_all = []
```

```python
thete = np.zeros((37*37,2))
```

```python
for i in range(37):
    for j in range(37):
        thete[i*37+j][0] = i*5
        thete[i*37+j][1] = j*5
```

```python
cx = b[:,0] + l*np.cos(thete*np.pi/180)
cy = b[:,1] + l*np.sin(thete*np.pi/180)
```

```python
len(cx)
```

```python
for i in range(len(thete)):
    center = sg.Point(cx[i][0], cy[i][0])
    radius = m[0]
    circle1 = sg.Circle(center, radius)
    
    center = sg.Point(cx[i][1], cy[i][1])
    radius = m[1]
    circle2 = sg.Circle(center, radius)
    
    result = sg.intersection(circle1,circle2 )
        
    if not result:
        continue
        
    if float(result[0].y) > float(result[1].y):
        p = [float(result[0].x), float(result[0].y)]
    else:
        p = [float(result[1].x), float(result[1].y)]
        
    if p[0]>-50 and p[0]<50 and p[1]>50 and p[1]<70:
        x.append(p[0])
        y.append(p[1])
        Thete1.append(thete[i][0])
        Thete2.append(thete[i][1])
        
    x_all.append(p[0])
    y_all.append(p[1])
    Thete1_all.append(thete[i][0])
    Thete2_all.append(thete[i][1])
        
    if thete[i][1] == 180:
        print("{}".format(thete[i][0]))
```
